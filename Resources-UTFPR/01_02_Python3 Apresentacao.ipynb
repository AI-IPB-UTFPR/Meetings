## Tratamento de Exceções

````
# exemplo de exceção
try:
    1 / 0
except Exception as e:
    print('ocorreu um erro', e)
finally:
  print('sempre executado, não obrigatório, interessante para limpar recursos')
````

````
#  exemplo com raise e else
try:
    raise 'Uma exceção qualquer'
except Exception as e:
    print('ocorreu um erro', e)
else:
  print('seria executado se desse tudo certo')
````

````
try:
    # Raise com tipo de erro definido
    raise IndexError("Exceção criada pq sim")
    
# trata um tipo de exceção
except IndexError as e:
    # pass faz com que nada seja exibido e o código siga
    pass
# multiplos tipos de erros podem ser tratados de uma só vez
except (TypeError, NameError):
    pass
else:
    print("Nenhum erro!")
finally:
    print("Finalmente acabou!")
````

### Variáveis e Métodos privados

Variáveis e métodos “privados”, que não podem ser acessados ​​exceto em métodos do próprio objeto não existem em Python. No entanto, existe uma convenção que é seguida pela maioria dos programas em Python: um nome prefixado com um sublinhado (por exemplo: _spam ) deve ser tratado como uma parte não-pública da API (seja ele uma função, um método ou um atributo de dados). Tais nomes devem ser considerados um detalhe de implementação e sujeito a alteração sem aviso prévio.

Uma vez que existe um caso de uso válido para a definição de atributos privados em classes (especificamente para evitar conflitos com nomes definidos em subclasses), existe um suporte limitado a identificadores privados em classes, chamado name mangling (literalmente: desfiguração de nomes). Qualquer identificador no formato `__spam` (no mínimo dois underscores _ no prefixo e no máximo um sufixo) é substituído por _nomeclasse__spam, onde nomeclasse é o nome da classe corrente (exceto quando o nome da classe é prefixado com um ou mais underscores _; nesse caso eles são omitidos). Essa desfiguração independe da posição sintática do identificador, desde que ele apareça dentro da definição de uma classe.

A desfiguração de nomes é útil para que subclasses possam sobrescrever métodos sem quebrar invocações de métodos dentro de outra classe. Por exemplo:




````
class Super_classe:
    def __init__(self, valores):
        self.items_list = []
        self.__update(valores) # referencia ao nome privado

    def update(self, valores): # parte da API, pode ser sobrescrito
        for item in valores:
            self.items_list.append(item)

    __update = update   # nome privado do método update

class Subclass( Super_classe):

    def update(self, keys, valores):
        # altera a assinatura de update()
        # mas não quebra o __init__() original
        for item in zip(keys, values):
            self.items_list.append(item)
````

###Herança Simples

A forma comum de herança é chamada de herança 
simples, na qual a nova classe é derivada de apenas 
uma classe já existente, porém é possível criar várias 
classes derivadas, criando uma hierarquia de classes.
Para localizar os métodos e atributos, a hierarquia é 
seguida de baixo para cima.

A nova  classe  pode  implementar  novos  métodos  e 
atributos  e  herdar  métodos  e  atributos  da  classe 
antiga  (que  também  pode  ter  herdado  de  classes 
anteriores), porém estes métodos e atributos podem 
ser substituídos na nova classe.

````
import time

class Mamifero():
  def __init__(self):
    self.qtdpatas = 4
    self.carnivoro = True
    
  def dormir(self,horas =3):
    print('Vou dormir por 3 segundos')
    time.sleep(horas)
    print('acordei')
    
class Cachorro(Mamifero):
  def __init__(self,cor,raca):
    
    ## falta algo ??
    #Mamifero.__init__(self)
    super().__init__()
    
    self.cor = cor 
    self.raca = raca;
  def dorme(self):
    self.dormir()
  def info(self):
    print('Quantidade de patas:',self.qtdpatas,'\n È carnivoro ? ',self.carnivoro,'\n qual a cor?',self.cor,'\n Qual a raça?',self.raca)
    
cao = Cachorro('cinza','Pastor Alemão')
cao.dormir()
cao.info()
    
 
````

### Herança Múltipla

Sim, Python suporta herança múltipla, Na herança múltipla, a nova classe deriva de duas ou 
mais classes já existentes.

````
#Mamifero.__init__(self)
#super().__init__()
import time
class Grande():
  def __init__(self):
    self.nervoso = True
    
  
class Mamifero():
  def __init__(self):
    self.qtdpatas = 4
    self.carnivoro = True
    
  def dormir(self,horas =3):
    print('Vou dormir por 3 segundos')
    time.sleep(horas)
    print('acordei')
    
class Cachorro(Mamifero,Grande):
  def __init__(self,cor,raca):
    ## falta algo ??
    
    super().__init__()
    
    Grande.__init__(self)
    
    self.cor = cor 
    self.raca = raca;
  def dorme(self):
    self.dormir()
  def info(self):
    print('Quantidade de patas:',self.qtdpatas,'È carnivoro ? ',self.carnivoro,'qual a cor?',self.cor,'Qual a raça?',self.raca, 'Está nervoso?',self.nervoso )
    
cao = Cachorro('cinza','Pastor Alemão')
cao.dormir()
cao.info()
    
 

````

## Mapeamento Objeto-Relacional do Python (ORM)

Mapeamento objeto-relacional (ou ORM, do inglês: Object-relational mapping) é uma técnica de desenvolvimento utilizada para reduzir a impedância da programação orientada a objetos utilizando bancos de dados relacionais. As tabelas do banco de dados são representadas através de classes e os registros de cada tabela são representados como instâncias das classes correspondentes.


### Peewee

Peewee é um ORM simples e pequeno. Tem poucos (mas expressivos) conceitos, facilitando o aprendizado e o uso intuitivo.

* Um ORM pequeno e expressivo
* Escrito em python com suporte para as versões 2.7+ e 3.4+ (desenvolvido com 3.6)
* Suporte embutido para SQLite, MySQL e Postgresql.
* numerosas extensões disponíveis (postgres hstore / json / arrays, pesquisa de texto completo em sqlite, migrações de esquema e muito mais).


[Tipos de dados suportados](http://docs.peewee-orm.com/en/latest/peewee/models.html)

````
#instalando o modulo
!pip install peewee
````

````
#Criando o banco de dados 
import peewee

db = peewee.SqliteDatabase('exemplo2.db')
````

````
class Usuario(peewee.Model):
    """
    Classe que representa a tabela Author
    """

    # A tabela possui apenas o campo 'name', que
    # receberá o nome do autor
    nome = peewee.CharField()
    email = peewee.CharField()
    senha = peewee.CharField()
    cpf = peewee.CharField()
    idade = peewee.IntegerField()
    
    

    class Meta:
        # Indica em qual banco de dados a tabela
        # 'Usuario' sera criada (obrigatorio). Neste caso,
        # utilizamos o banco 'exemplo.db' criado anteriormente.
        database = db
        
Usuario.create_table()  #criar tabela

alberto = Usuario.create(nome='Alberto da Silva',email = 'albertosilva@grr.la',senha='1234',cpf='097.078.034-21',idade = 21 )   
print(alberto.nome)

joao = Usuario.create(nome='João da Silva',email = 'joao@grr.la',senha='3214',cpf='097.079.034-21',idade = 27)   
print(joao.nome)

````

````
#get all tables sqlite3:SELECT * FROM sqlite_master where type='table';
import sqlite3
conn = sqlite3.connect('exemplo2.db')
cursor = conn.cursor()
cursor.execute("""
   SELECT * FROM usuario;
    """)
print(cursor.fetchall())
````
